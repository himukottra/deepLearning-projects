```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, Dense, Flatten, Dropout
from tensorflow.keras.utils import to_categorical

# If your target is integer encoded
y_train_cat = to_categorical(y_train, num_classes=3)
y_test_cat = to_categorical(y_test, num_classes=3)

# Reshape X to (samples, timesteps, channels) for Conv1D
X_train_cnn = X_train_scaled.reshape((X_train_scaled.shape[0], X_train_scaled.shape[1], 1))
X_test_cnn  = X_test_scaled.reshape((X_test_scaled.shape[0], X_test_scaled.shape[1], 1))

model = Sequential()

# 1D Convolutional layer
model.add(Conv1D(filters=32, kernel_size=3, activation='relu', input_shape=(20, 1)))
model.add(Dropout(0.2))  # prevent overfitting

# Second Conv1D layer
model.add(Conv1D(filters=64, kernel_size=3, activation='relu'))
model.add(Dropout(0.2))

# Flatten and connect to Dense layers
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dense(3, activation='softmax'))  # 3-class classification

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train
history = model.fit(X_train_cnn, y_train_cat, epochs=50, batch_size=32,
                    validation_data=(X_test_cnn, y_test_cat))
```

```python
# For one-hot encoded y
y_pred_prob = model.predict(X_test_cnn)      # shape: (num_samples, 3)
y_pred_classes = y_pred_prob.argmax(axis=1)  # pick the class with highest probability
```

```python
y_true_classes = y_test_cat.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score, classification_report

acc = accuracy_score(y_true_classes, y_pred_classes)
print("Test Accuracy:", acc)

# Optional: detailed metrics
print(classification_report(y_true_classes, y_pred_classes))
```
